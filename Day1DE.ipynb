```python
import matplotlib.pyplot as plt
```

```python
months=['jan','feb','march','April','May','June']
traffic=[150,200,100,220,250,210]

plt.plot(months,traffic)
# plt.show()
plt.xlabel('Month')
plt.ylabel('Monthly traffic')
plt.show()
```

```python
plt.plot(months,traffic,marker='^',markersize=10,color='green',linestyle='-.',linewidth=3)
plt.xlabel('Month')
plt.ylabel('Monthly traffic')
plt.show()
```

```python
months=['jan','feb','march','April','May','June']
producta_revenue=[45,55,40,70,80,75]
productb_revenue=[35,40,50,55,70,68]

plt.plot(months,producta_revenue,marker='o',markersize=10,color='blue',linestyle='-',label='Product A')
plt.plot(months,productb_revenue,marker='s',markersize=10,color='red',linestyle='--',label='Product B')

plt.xlabel('Months')
plt.ylabel('Monthly revenue')
plt.title('Revenue comparison')

plt.legend()
plt.show()
```
