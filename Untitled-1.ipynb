import unittest

class TestCalc(unittest.TestCase):

    def test_add(self):
        result, success = calculator(2, 3, 1)
        self.assertTrue(success)
        self.assertEqual(result, 5)

    def test_subtract(self):
        result, success = calculator(5, 3, 2)
        self.assertTrue(success)
        self.assertEqual(result, 2)

    def test_multiply(self):
        result, success = calculator(2, 4, 3)
        self.assertTrue(success)
        self.assertEqual(result, 8)

    def test_divide(self):
        result, success = calculator(10, 2, 4)
        self.assertTrue(success)
        self.assertEqual(result, 5)

    def test_divide_by_zero(self):
        result, success = calculator(10, 1, 4)
        self.assertFalse(success)
        self.assertIsNone(result,10)

    def test_invalid_case(self):
        result, success = calculator(2, 3, 99)
        self.assertFalse(success)
        self.assertIsNone(result)


if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
