```python
def calculator(a, b, case):
    try:
        match case:
            case 1:
                return a + b
            case 2:
                return a - b
            case 3:
                return a * b
            case 4:
                try:
                    return a / b
                except ZeroDivisionError as e:
                    print(e)
    except case:
        return ValueError
a=int(input("Enter the first number"))
b=int(input("Enter the second number"))
print("Enter 1 for addition \nEnter 2 for substraction \nEnter 3 for multiplication \nEnter 4 for division")
case=int(input())
# case=int(input("Enter 1 for addition \nEnter 2 for substraction \nEnter 3 for multiplication \nEnter 4 for division"))
print(calculator(a,b,case))
    
```
